```python
import numpy as np
import matplotlib.pyplot as plt
import json
from trimesh import *

X, _, _ = load_off("tr_reg_090.off")

weights = json.load(open("weights.json"))
n_weights = len(weights)
n_vertices = 1 + max([ max([int(k) for k in value.keys()]) for value in weights.values()])
W = np.zeros((n_vertices, n_weights))
key2idx = {key:i for i, key in enumerate(weights.keys())}
labels = []
for j, (key, value) in enumerate(weights.items()):
    labels.append(key)
    for i, w in value.items():
        i = int(i)
        W[i, j] = w

idx = 0
plt.scatter(X[:, 0], X[:, 1], s=1, c=np.sum(W > 0, axis=1))
plt.axis("equal")
plt.title(labels[idx])
plt.colorbar()
```
